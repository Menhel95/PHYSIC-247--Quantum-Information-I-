# Exercise 1

```python
import numpy as np
import matplotlib.pyplot as plt
shots = [10, 100, 1000, 10000]
results = []
for shot in shots:
    # Generate random numbers between 1 and 6 for die rolls
    data = np.random.randint(1, 7, shot)
    results.append(data)
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))
axes = axes.flatten()

# Create a histogram for each set of shots
for i, data in enumerate(results):
    axes[i].hist(data, bins=6, range=(1, 7), alpha=0.7, color='blue', edgecolor='black')
    axes[i].set_title(f'Histogram: {shots[i]} shots')
    axes[i].set_xlabel('Outcome')
    axes[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

# Exercise 2

```python
# Import necessary libraries
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram
from numpy import pi

# Create a quantum circuit with one qubit
qc = QuantumCircuit(1)

# Apply a Hadamard gate to create a superposition
qc.h(0)

# Apply an S-dagger (inverse phase) gate to introduce the negative imaginary component
qc.sdg(0)

# Draw the circuit
qc.draw('mpl')

# Simulate the circuit
simulator = Aer.get_backend('statevector_simulator')
result = execute(qc, simulator).result()
statevector = result.get_statevector()

# Print the resulting state vector
print("State vector:", statevector)

# Simulate and plot histogram
simulator = Aer.get_backend('qasm_simulator')
qc.measure_all()
result = execute(qc, simulator, shots=1024).result()
counts = result.get_counts()

# Plot the histogram
plot_histogram(counts)
```
