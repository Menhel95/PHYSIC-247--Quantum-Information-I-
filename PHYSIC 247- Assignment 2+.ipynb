# Exercise 1

```python
import numpy as np
import matplotlib.pyplot as plt
shots = [10, 100, 1000, 10000]
results = []
for shot in shots:
    # Generate random numbers between 1 and 6 for die rolls
    data = np.random.randint(1, 7, shot)
    results.append(data)
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))
axes = axes.flatten()

# Create a histogram for each set of shots
for i, data in enumerate(results):
    axes[i].hist(data, bins=6, range=(1, 7), alpha=0.7, color='blue', edgecolor='black')
    axes[i].set_title(f'Histogram: {shots[i]} shots')
    axes[i].set_xlabel('Outcome')
    axes[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

# Exercise 2

```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_histogram

# Create quantum and classical registers
n_qubit = 1
qc2 = QuantumCircuit(n_qubit, n_qubit)

# Apply Hadamard gate
qc2.h(0)

# Apply phase shift (S gate is a phase shift of -/2)
qc2.sdg(0)

# Measure the circuit
qc2.measure(0, 0)

# Draw the circuit
qc2.draw('mpl')
```

```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
# |- state circuit
n_qubit = 2
qr = QuantumRegister(n_qubit)
cr = ClassicalRegister(n_qubit)
qc = QuantumCircuit(qr, cr)

qc.h(qr[0])
qc.cx(qr[0], qr[1])
qc.z(qr[1])

qc.draw('mpl')
```
